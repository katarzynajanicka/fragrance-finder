## 2. Similar fragrances finder

Fuzzywuzzy is a Python library that uses the Levenshtein distance to calculate the differences between different strings.
- <b> fuzz.ratio </b> compares the entire string similarity in order.
- <b> fuzz.partial_ratio </b> compares partial string similarity.
- <b> fuzz.token_sort_ratio </b> ignores word order. <br>
It tokenizes strings and clears them by returning lowercase, removing punctuation marks, and sorting them alphabetically. Then finds the Levenshtein distance and returns the similarity percentage.
- <b> fuzz.token_set_ratio </b> is similar to fuzz.token_sort_ratio but ignores duplicated words.

### 2.1. Angel Mugler for women

```python
# select information on the fragrance
df.iloc[[7827]]
```

```python
# find most similar fragrances based on the accords
angel = 'sweet,patchouli,warm spicy,fruity,caramel'

df_angel_ = df[df.apply(lambda row: fuzz.token_sort_ratio(row['accords_'], angel), axis=1)>88]
df_angel_
```

```python
# calculate the similarity percentage between the strings
df_angel = df_angel_.copy()
df_angel['similarity']= df_angel.apply(lambda row: fuzz.ratio(row['accords_'], angel), axis=1)

# define violet color palette
violet=sns.color_palette('light:mediumvioletred', as_cmap=True)

# color the similarity column based on its values
df_angel.sort_values(by='similarity', ascending=False).style.background_gradient(subset='similarity', cmap=violet)
```
